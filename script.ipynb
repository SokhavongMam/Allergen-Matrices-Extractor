```python
import os
import csv
import pandas as pd
```

```python
def convert_sentence(sentence):
    
    # Split the sentence into words
    words = sentence.split()
    # Convert each word to lowercase and capitalize the first letter
    converted_words = [word[:1].upper() + word[1:].lower() for word in words]    
    # Join the words back into a sentence
    converted_sentence = ' '.join(converted_words)
    
    
    return converted_sentence

def import_and_trim(file):

    df = pd.read_excel(file, usecols="B:ZZ", skiprows=[0, 1, 2, 3, 4, 5, 6, 8, 9])

    


    for column in df.columns:
        if column == "CRUSTACEAN\n(SHELLFISH) ":
            df = df.rename(columns={column: "Shellfish"})
        elif column == "FISH OR MOLLUSC\n(SHELLFISH) ":
            df = df.rename(columns={column: "Fish or Mollusc"})
        elif column == "GLUTEN OR\nWHEAT ":
            df = df.rename(columns={column: "Gluten"})
        elif column == "LACTOSE\n(MILKPRODUCTS) ":
            df = df.rename(columns={column: "Lactose"})
        else:
            df = df.rename(columns={column: convert_sentence(column)})
    return df
```

```python
def extract_and_write(df):

    grp_title = df.columns[0]


    # iterrows iterates through all rows
    # row[0]: name of food item


    with open("./result.csv", 'a', newline='') as file:
        writer = csv.writer(file)


        for index, row in df.iterrows(): # iterates through all row in food item
            tmp = []
            for column in range(2, len(row)): # iterates through all column in row, skip food name and "no allergen"
                if row[column] == "âˆš":
                    tmp.append(df.columns[column]) # add the column of the allergens[index of column]
                    # print(row[column], df.columns[column])

            # print(type(row[0]))

            if type(row[0]) is float:
                break
            
            food_name = convert_sentence(row[0])

            # if food_name == None:
            #     break

            if tmp == []:
                diet_name = None
            else:
                diet_name = f"Contains: {', '.join(tmp)}"

            writer.writerow([food_name, diet_name, grp_title])
```

```python
files = os.listdir("./files")

# files = ["Copy of Record 11 - Allergies matrix - Curate your plate (Pasta dish).xlsx"]

for file in files:
    # print(str(f"./files/{file}"))
    dataframe = import_and_trim(f"./files/{file}")
    extract_and_write(dataframe)
    # break
```
